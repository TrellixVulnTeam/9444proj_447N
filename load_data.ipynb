```python
# use to check if image matches
def show_img_person(name):
    name = name.replace(' ', '_')
    img_dir = os.path.join("lfw_funneled", name)
    img_path = os.path.join("lfw_funneled", name, os.listdir(img_dir)[0])
    img = mpimg.imread(img_path)
    imgplot = plt.imshow(img)
    plt.show()
```
